```python
plt.figure(figsize=(10,10))
values =[326,326,327,334,335]
labels =[0.23,0.21,0.23,0.29,0.31]
explode = (0, 0, 0, 0, 0.1)
plt.pie(values, labels= labels,explode=explode,counterclock=False,shadow=True,autopct='%0.2f%%')
plt.title('Caret with MRP')
plt.legend(labels,loc=0) 
plt.show()
```
